```python
import pandas as pd
import numpy as np
```

```python
paris_listings = pd.read_csv('paris_airbnb.csv')
paris_listings.iloc[0]
```

### Distance Euclidienne ###

```python
# numpy.abs()
```

```python
our_acc_value = 3 #variable pour le nb de personne que peut accueillir notre logement

first_living_space_value = paris_listings.iloc[0]['accommodates'] #on prend la première ligne et la valeur de la colonne accomodates

first_distance = np.abs(first_living_space_value - our_acc_value) #methode .abs de numpy - Calculate the absolute value element-wise
print(first_distance)
```

### Calculer la distance pour toutes les observations ###

```python
# df['accommodates'].apply(lambda x: x + 2) #methode lambda pour assigner +2 à chaque valeur de x
```

```python
new_listing = 3
paris_listings['distance'] = paris_listings['accommodates'].apply(lambda x: np.abs(x - new_listing)) #on créé une nouvelle colonne distance avec la formule
print(paris_listings['distance'].value_counts()) #.value_counts() Return a Series containing counts of unique values
```

```python
print(paris_listings['distance'])
```

### Randomisation et triage ###

```python
paris_listings[paris_listings['distance'] == 0]['accommodates']
```

```python
# np.random.permutation(len(paris_listings))
```

```python
np.random.seed(1)
paris_listings = paris_listings.loc[np.random.permutation(len(paris_listings))] #on randomise l'ensemble des données
paris_listings = paris_listings.sort_values('distance') #on tri sur la colonne distance
paris_listings.iloc[0:10]['price'] #on affiche les 10 premières valeurs de la colonne price
```

### Prix moyen ###

```python
# mean() fonction pour calculer la moyenne
```

```python
stripped_commas = paris_listings['price'].str.replace(',', '') #ici on remplace les , par rien
stripped_dollars = stripped_commas.str.replace('$', '') #ici on remplace les $ par rien
paris_listings['price'] = stripped_dollars.astype('float') #on transforme le type de la colonne price en Float
mean_price = paris_listings.iloc[0:5]['price'].mean() #on calcule la moyenne des prix sur les 5 premières colonnes
print(mean_price)
```
