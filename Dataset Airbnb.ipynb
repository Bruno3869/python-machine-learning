```python
import numpy as np
import pandas as pd
np.random.seed(1)

paris_listings = pd.read_csv('paris_airbnb.csv')
stripped_commas = paris_listings['price'].str.replace(',', '') #ici on remplace les , par rien
stripped_dollars = stripped_commas.str.replace('$', '') #ici on remplace les $ par rien
paris_listings['price'] = stripped_dollars.astype('float') #on transforme le type de la colonne price en Float
paris_listings = paris_listings.loc[np.random.permutation(len(paris_listings))] #on randomise l'ensemble des données

def predict_price(new_listing): #on définie la fonction predict_price
    temp_df = paris_listings.copy() #on copie le dataset dans une nouvelle variable
    temp_df['distance'] = temp_df['accommodates'].apply(lambda x: np.abs(x - new_listing)) #on créé une nouvelle colonne distance avec la formule
    temp_df = temp_df.sort_values('distance') #on tri sur la colonne distance
    nearest_neighbors = temp_df.iloc[0:5]['price'] #on assigne à la variable nearest_neighbors les 5 premières valeurs
    predicted_price = nearest_neighbors.mean() #on assigne à la variable predicted_price la moyenne
    return(predicted_price)

acc_one = predict_price(1) #la prévision avec 1 chambre
acc_two = predict_price(2) #la prévision avec 2 chambres
acc_four = predict_price(4) #la prévision avec 3 chambres

print(acc_one)
print(acc_two)
print(acc_four)
```

```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
np.random.seed(1)

paris_listings = pd.read_csv('paris_airbnb.csv')
stripped_commas = paris_listings['price'].str.replace(',', '') #ici on remplace les , par rien
stripped_dollars = stripped_commas.str.replace('$', '') #ici on remplace les $ par rien
paris_listings['price'] = stripped_dollars.astype('float') #on transforme le type de la colonne price en Float
paris_listings = paris_listings.loc[np.random.permutation(len(paris_listings))] #on randomise l'ensemble des données
train_df = paris_listings.iloc[0:6000] #on divise le dataset en 2
test_df = paris_listings.iloc[6000:] #on divise le dataset en 2

def predict_price(new_listing): #on définie la fonction predict_price
    temp_df = train_df.copy() #on copie le dataset dans une nouvelle variable
    temp_df['distance'] = temp_df['accommodates'].apply(lambda x: np.abs(x - new_listing)) #on créé une nouvelle colonne distance avec la formule
    temp_df = temp_df.sort_values('distance') #on tri sur la colonne distance
    nearest_neighbors = temp_df.iloc[0:5]['price'] #on assigne à la variable nearest_neighbors les 5 premières valeurs
    predicted_price = nearest_neighbors.mean() #on assigne à la variable predicted_price la moyenne
    return(predicted_price)
```
