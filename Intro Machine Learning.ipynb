```python
#retourner une liste de toutes les colonnes float et l'attribuer à la variable airbnb_cols
airbnb_cols = airbnb.select_dtypes(include=['float']).columns.to_list()
```

```python
for col in airbnb_cols:
    print(col)
```

```python
#retourner les moyennes des colonnes float
for col in airbnb_cols:
    calc_mean = np.mean(airbnb[col])
    print('moyenne de la colonne ' + col + ':', calc_mean)
```

```python
#fonction pour diviser en 2 le dataframe
def split_func(df):
    df1 = df.loc[0:len(df)/2] #1er dataframe de 0 à la moitiè
    df2 = df.loc[len(df)/2:]  #2ème dataframe de la moitié à la fin
    return(df1, df2)

airbnb1, airbnb2 = split_func(airbnb) #On divise en 2 le dataframe
```

```python
airbnb1.info()
```

```python
#fonction pour ajouter 100 à toutes les colonnes
def add_hundred(x):
    return x+100

airbnb[airbnb_cols].apply(add_hundred)
```

```python
#on utilise la fonction lambda pour ajouter 300 à chaque colonne
airbnb[airbnb_cols].apply(lambda x: x+300)
```
